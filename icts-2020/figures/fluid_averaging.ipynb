A sketch of averaging.

```python
import numpy
from matplotlib import pyplot
```

```python
n_p = 50

positions_blue = numpy.random.rand(n_p, 2)
positions_red = numpy.random.rand(n_p, 2)
positions_green = numpy.random.rand(n_p, 2)

vels_blue = numpy.array([0.2, 0.5]) + 0.2*(numpy.random.rand(n_p, 2) - 0.5)
vels_red = numpy.array([0.2, 0.5]) + 0.2*(numpy.random.rand(n_p, 2) - 0.5)
vels_green = numpy.array([0.2, 0.5]) + 0.2*(numpy.random.rand(n_p, 2) - 0.5)
```

```python
fig, axes = pyplot.subplots(1, 2, figsize=(14, 7))
axes[0].scatter(positions_blue[:, 0], positions_blue[:, 1], color='b', s=60)
axes[0].scatter(positions_red[:, 0], positions_red[:, 1], color='r', s=60)
axes[0].scatter(positions_green[:, 0], positions_green[:, 1], color='g', s=60)

axes[0].quiver(positions_blue[:, 0], positions_blue[:, 1], vels_blue[:, 0], vels_blue[:, 1], color='b', scale=7)
axes[0].quiver(positions_red[:, 0], positions_red[:, 1], vels_red[:, 0], vels_red[:, 1], color='r', scale=7)
axes[0].quiver(positions_green[:, 0], positions_green[:, 1], vels_green[:, 0], vels_green[:, 1], color='g', scale=7)

axes[1].scatter([0.5], [0.5], color='k', s=200)
axes[1].quiver([0.5], [0.5], [0.2], [0.5], color='k', scale=2, width=0.01, headwidth=4)

for ax in axes:
    ax.set_xlim(-0.01, 1.1)
    ax.set_ylim(-0.01, 1.1)
    ax.set_aspect('equal', 'box')
#     ax.axis('off')
    ax.set_xticks([])
    ax.set_yticks([])
fig.tight_layout()
# fig.savefig("fluid_averaging.png", bbox_inches="tight")
```
