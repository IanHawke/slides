<section class="section3"><h3>Packages</h3>
<p>As your document grows in size you will want to improve its appearance and to use shortcuts for doing more complicated tasks.</p>
<p>External packages may alter the appearance of the document, or provide useful commands.
This is similar to packages or modules in Python.
To include an external package then, in the preamble, use the command <code>\usepackage{...}</code>.
For example, the command <code>\usepackage{amsmath}</code> imports the standard typesetting of the American Mathematical Society, which is particularly useful for mathematicians.
A brief list of useful packages includes</p>
<ul>
<li><code>graphicx</code>: include graphics files for figures in your document.</li>
<li><code>amsmath</code>, <code>amssymb</code>, <code>amsthm</code>: mathematical typesetting.</li>
<li><code>hyperref</code>: add internal and external hyperlinks to the PDF document.</li>
<li><code>fancyhdr</code>, <code>enumerate</code>, <code>caption</code>, <code>geometry</code>: control the document layout and appearance more transparently.</li>
<li><code>physics</code>: helps with, e.g., differential equations.</li>
</ul>

<li>Look at the <code>BasicLatexFigure.tex</code> document.</li>

</ul>
